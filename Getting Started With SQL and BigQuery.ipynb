```python
#To access the dataset,
# Construct a reference to the "hacker_news" dataset
dataset_ref = client.dataset("hacker_news", project="bigquery-public-data")
```

```python
# List all the tables in the "hacker_news" dataset
tables = list(client.list_tables(dataset))

# Print names of all tables in the dataset (there are four!)
for table in tables:
    print(table.table_id)
```

```python
# Construct a reference to the "full" table
table_ref = dataset_ref.table("full")

# API request - fetch the table
table = client.get_table(table_ref)
```

 **Table Schema**--
The structure of a table is called its schema

```python
# Print information on all the columns in the "full" table in the "hacker_news" dataset
table.schema
```

SchemaField('title', 'STRING', 'NULLABLE', 'Story title', (), None),



| Part            | Meaning                                                                           |
| --------------- | --------------------------------------------------------------------------------- |
| `'title'`       | **Column name** → the name of the field in the table                              |
| `'STRING'`      | **Data type** → this column stores text (strings)                                 |
| `'NULLABLE'`    | **Mode** → this field can have missing values (`NULL`)                            |
| `'Story title'` | **Description** → human-readable explanation of what the field means              |
| `()`            | **Field subfields** → used only for nested records (this is empty, so not nested) |
| `None`          | **Policy tag** → used for data classification (e.g., for DLP), here it's not set  |
